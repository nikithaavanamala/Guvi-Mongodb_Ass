# Sum

```python
collect.aggregate([{
  '$group': {
    '$objects': {'$push': 'name'},
  },
}, {
  '$aggregate': {
    'scores': 1
     {
        '$unwind': '$scores'
    },
    "$sum":{"$scores.score","$scores.type" == "exam"},
  }
}])
```

# Average

```python
collect.aggregate([{
  '$group': {
    '$objects': {'$push': 'name'},
  },
}, {
  '$aggregate': {
    'scores': 1
     {
        '$unwind': '$scores'
    },
    "$avg":{"$scores.score","$scores.type" == "exam"},
  }
}])
```
